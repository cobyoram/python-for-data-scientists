```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.random import randint
from numpy import mean
```

# Gamma distribution

### Law of large numbers

```python
results = []
for num_trials in range(1, 1000): 
    trials = np.random.gamma(5, 2, num_trials)
    mean_of_trials = trials.mean()
    results.append(mean_of_trials)
    
df = pd.DataFrame({'trials': results})

df.plot()
```

### Central limit theorem

```python
means = [mean(np.random.gamma(2, 5, 1000)) for _ in range(1000)]

plt.hist(means)
plt.show()
```

# Binomial

### Law of large numbers

```python
results = []
for num_trials in range(1, 1000):
    trials = np.random.randint(1, 21, num_trials)
    mean_of_trials = trials.mean()
    results.append(mean_of_trials)
    
df = pd.DataFrame({'trials': results})

df.plot()
```

### Central limit theorem

```python
means = [mean(np.random.randint(1, 21, 1000)) for _ in range(1000)]

plt.hist(means)
plt.show()
```
